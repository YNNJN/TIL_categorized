```python
#9012_괄호
#스택이용
def find(data):
    s = []
    for b in data:
        if len(s) == 0 and b == ')':
            return 'NO'
        if b == '(':
            s.append(b)
        else:
            if s[-1] == '(':
                s.pop()
            else:
                return 'NO'
    if len(s) == 0:
        return 'YES'
    return 'NO'

for tc in range(int(r())):
    data = r().strip()
    print(find(data))
```

**큐**는 스택의 연산과 더불어 다음의 연산을 추가로 가짐
- front: 큐의 가장 앞에 있는 자료를 보는 연산
- back: 큐의 가장 뒤에 있는 자료를 보는 연산
- 그리고 헉 큐의 pop은 가장 앞의 값을 꺼내어 출력함

```python
def push(n):
    q.append(n)

def pop():
    if not q:
        print(-1)
    else:
        print(q[0])
        del(q[0])

def size():
    print(len(q))

def empty():
    if not q:
        print(1)
    else:
        print(0)

def front():
    if not q:
        print(-1)
    else:
        print(q[0])

def back():
    if not q:
        print(-1)
    else:
        print(q[-1])

n = int(input())
q = []

for _ in range(n):
    cmd = list(input().split())

    if cmd[0] == 'push':
        push(cmd[1])
    elif cmd[0] == 'pop':
        pop()
    elif cmd[0] == 'size':
        size()
    elif cmd[0] == 'empty':
        empty()
    elif cmd[0] == 'front':
        front()
    elif cmd[0] == 'back':
        back()
```

**덱**은 양 끝에서만 자료를 넣고 뺄 수 있는 자료구조
- Deque, Double-ended queue
- 큐의 연산과 더불어 다음의 연산을 추가로 가짐
    - push_front: 큐의 앞에 자료를 넣는 연산
    - push_back: 큐의 뒤에 자료를 넣는 연산
    - pop_front: 큐의 앞에서 자료를 빼는 연산
    - pop_back: 큐의 뒤에서 자료를 빼는 연산

```python
#10866_덱
import sys
r = sys.stdin.readline

def push_front(n):
    d.insert(0, n)
    
def push_back(n):
    d.append(n)
    
def pop_front():
    if not d:
        print(-1)
    else:
        print(d[0])
        del(d[0])
        
def pop_back():
    if not d:
        print(-1)
    else:
        print(d[-1])
        del(d[-1])

def size():
    print(len(d))

def empty():
    if not d:
        print(1)
    else:
        print(0)

def front():
    if not d:
        print(-1)
    else:
        print(d[0])

def back():
    if not d:
        print(-1)
    else:
        print(d[-1])

n = int(r())
d = []

for _ in range(n):
    cmd = list(r().split())

    if cmd[0] == 'push_front':
        push_front(cmd[1])
    elif cmd[0] == 'push_back':
        push_back(cmd[1])
    elif cmd[0] == 'pop_front':
        pop_front()
    elif cmd[0] == 'pop_back':
        pop_back()
    elif cmd[0] == 'size':
        size()
    elif cmd[0] == 'empty':
        empty()
    elif cmd[0] == 'front':
        front()
    elif cmd[0] == 'back':
        back()
```

```python
#2644_촌수계산
#트리
'''
촌수 계산의 핵심은 나와 가장 공통된 조상을 찾는 것, LCA
n <= 100 의 경우

'''

n = int(input())
a, b = map(int, input().split())
p = [0 for _ in range(n + 1)]

for _ in range(int(input())):
    x, y = map(int, input().split())
    p[y] = x  # y의 부모는 x

aa, ba = [], []  # 조상리스트
ad, bd = 0, 0  # 촌수

while p[a]:
    aa.append((a, ad))
    ad += 1
    a = p[a]
    
while p[b]:
    ba.append((b, bd))
    bd += 1
    b = p[b]

for i in aa:
    for j in ba:
        if i[0] == j[0]:
            print(i[1] + j[1])
            exit()
print(-1)
```

- 이진트리를 활용한 자료구조, heap과 bst
- **Heap**이란 우선순위가 가장 높은 값을 root node로 만듦
    - C에서는 Priority queue라고 부름, 즉 우선순위 큐
        - pop 기능이 있어 -> 1. 정렬 가능(Heap sort, 이진트리의 성격을 지녀, 항상 stable하게 log n의 시간복잡도를 지님)
    - 3 2 9 1 4 5
    - 큰 값에 우선순위를 두어, 부모보다 자식이 크면 둘을 자리 바꿈
    - 최대값이 가장 꼭대기로 올라오게 됨
    - 시간복잡도는 다음과 같음
        - n층이 있을 때 2^(n+1)-1의 원소가 있음
        - 2^n = k 라고 하면, 트리의 높이는 log_2(k)
        - 최대값 갱신에는 원소 k, 높이 logk -> O(NlogN)
        - 즉 O(NlogN)으로 힙을 정렬할 수 있음
- **BST**란 이진검색트리 (Binary Search Tree)
    - 루트 노드와의 크기 비교로 가지를 쳐나감
    - 값을 찾기 쉬운 특성
        - 검색하는 시간 복잡도가 logN (이진트리의 높이)이니
        - 그러나 최악의 경우, 즉 한쪽으로 치우친 트리의 경우
            - 메모리는 더 들고 리스트에서 찾는 것과 다를 바 없게 됨
    - 때문에 balancing factor를 이용
        - AVL, Splan tree, Red-black tree 등 BST의 다수 변형이 존재
    - 파이썬에서는 set, dict -> hash 배열로 매핑하여 사용, set과 dict도 bst로 만든 것
        - 집합의 원소 유/무를 판별하는 것이니 binary tree 구조 접근이 적합
